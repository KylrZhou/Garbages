```python
class node():
    def __init__(self, next, prob, reward):
        self.next = next
        self.reward = reward
        self.prob = prob
        self.visit = False
        self.result = 0
```

```python
n = []
n.append(node([1, 4], [0.8, 0.2], [2, 1]))
n.append(node([2, 4], [0.9, 0.1], [-5, 1]))
n.append(node([3, 4, 5], [0.4, 0.2, 0.4], [100, 2, 4]))
n.append(node([], [], []))
n.append(node([1, 4], [0.4, 0.6], [1, -3]))
n.append(node([0, 5], [0.1, 0.9], [1, -10]))
```

```python
def recursive_func(x):
    print(f"Recursed to {x}")
    n[x].visit = True
    bst = 0
    flag = None
    for idx, i in enumerate(n[x].next):
        if n[i].visit == False:
            print(f"    Visiting {i}")
            tmp = recursive_func(i) + (n[x].prob[idx] * n[x].reward[idx])
        else:
            print(f"    Already visited and best score of {i} is {n[i].result}")
            tmp = n[i].result
    
        print(f"        Reward from {x} to {i} is {tmp} current best is {bst}")
        if tmp > bst:
            print(f"        Replaced by new best score {tmp}")
            bst = tmp
            flag = idx
    if flag is not None:
        print(f"Final next step for {x} is {n[x].next[flag]}")
        n[x].result = bst    
    return n[x].result
```

```python
recursive_func(0)
```

```python
for i in n:
    print(i.result)
```

```python
class node():
    def __init__(self, next, prob, reward):
        self.next = next
        self.reward = reward
        self.prob = prob
        self.visit = []
        for i in range(len(next)):
            self.visit.append(False)
        self.result = 0
```

```python
n = []
n.append(node([5], [0.1], [1]))
n.append(node([0, 4], [0.8, 0.4], [2, 1]))
n.append(node([1], [0.9], [-5]))
n.append(node([2], [0.4], [100]))
n.append(node([0, 1, 2, 4], [0.2, 0.1, 0.2, 0.6], [1, 1, 2, -3]))
n.append(node([2, 5], [0.4, 0.9], [5, -10]))
```

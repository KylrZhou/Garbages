```python
import os
import PIL.Image as Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('fig.jpg')
print(img.size)
plt.imshow(img)
plt.show()
```

```python
def ImgsizeAdjust(input_img, target_size, bbox_offsets = False):
    img_aspect_ratio = input_img.size[0] / input_img.size[1]
    back_ground = Image.new(input_img.mode, (target_size[0], target_size[1]))
    if img_aspect_ratio > 0.75:
        input_img = input_img.resize((target_size[0], int(target_size[0] / img_aspect_ratio)))
        print(target_size[1], input_img.size[1])
        offsets = (0, int((target_size[1] - input_img.size[1]) / 2) )
        back_ground.paste(input_img, offsets)
    else:
        input_img = input_img.resize((int(target_size[1] * img_aspect_ratio), target_size[1]))
        print(target_size[0], input_img.size[0])
        offsets = (int((target_size[0] - input_img.size[0]) / 2) ,target_size[1])
        back_ground.paste(input_img, offsets)
    if bbox_offsets:
        return back_ground, offsets
    else:
        return back_ground
```

```python
oft = 0
img, oft = ImgsizeAdjust(img, (192,256), True)
```

```python
plt.imshow(img)
plt.show()
```
